# <span style='color:Black'> Correlation matrix Visualization </span>

```python
import numpy as np
import pandas as pd
import seaborn as sns

data = pd.read_csv('../data/combined_csv_missingval_mean.csv',encoding='iso-8859-1')
data=data.drop(['STATION CODE'], axis = 1) 
corr = data.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap = sns.diverging_palette(20, 220, n=200),
    square=True
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```
